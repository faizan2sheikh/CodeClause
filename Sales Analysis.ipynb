```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sales_df = pd.read_csv('vgsales.csv')
```

```python
sales_df.head()
```

```python
plt.figure(figsize=(20,10))
most_sales=sales_df.groupby('Publisher')[['Global_Sales']].sum().sort_values(by='Global_Sales',ascending=False).head(10)
sns.barplot(x=most_sales.index,y='Global_Sales',data=most_sales)
```

```python
plt.figure(figsize=(20,10))
most_sales=sales_df.groupby('Genre')[['Global_Sales']].sum().sort_values(by='Global_Sales',ascending=False).head(10)
sns.barplot(x=most_sales.index,y='Global_Sales',data=most_sales)
```

```python
plt.figure(figsize=(20,10))
most_sales=sales_df.groupby('Platform')[['Global_Sales']].sum().sort_values(by='Global_Sales',ascending=False).head(10)
sns.barplot(x=most_sales.index,y='Global_Sales',data=most_sales)
```

```python
sales_df.head(20)
```

```python
fig, axes = plt. subplots(2, 2, figsize=(18,8))
most_sales=sales_df.groupby('Genre')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()
sns.barplot(x=most_sales.index,y='NA_Sales',data=most_sales,ax=axes[0,0],order=most_sales.sort_values('NA_Sales',ascending = False).index)
sns.barplot(x=most_sales.index,y='EU_Sales',data=most_sales,ax=axes[0,1],order=most_sales.sort_values('EU_Sales',ascending = False).index)
sns.barplot(x=most_sales.index,y='JP_Sales',data=most_sales,ax=axes[1,0],order=most_sales.sort_values('JP_Sales',ascending = False).index)
sns.barplot(x=most_sales.index,y='Other_Sales',data=most_sales,ax=axes[1,1],order=most_sales.sort_values('Other_Sales',ascending = False).index)
fig.tight_layout()
```

```python
fig, axes = plt. subplots(2, 2, figsize=(18,8))

most_sales=sales_df.groupby('Publisher')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum()
sns.barplot(x=most_sales.index,y='NA_Sales',data=most_sales,ax=axes[0,0],order=most_sales.sort_values('NA_Sales',ascending = False).index[:5])
sns.barplot(x=most_sales.index,y='EU_Sales',data=most_sales,ax=axes[0,1],order=most_sales.sort_values('EU_Sales',ascending = False).index[:5])
sns.barplot(x=most_sales.index,y='JP_Sales',data=most_sales,ax=axes[1,0],order=most_sales.sort_values('JP_Sales',ascending = False).index[:5])
sns.barplot(x=most_sales.index,y='Other_Sales',data=most_sales,ax=axes[1,1],order=most_sales.sort_values('Other_Sales',ascending = False).index[:5])
fig.tight_layout()
```

```python
plt.figure(figsize=(10,8))
act = sales_df[sales_df['Publisher'] == 'Nintendo'].groupby('Year')[['Global_Sales']].mean()
plt.plot(act['Global_Sales'], linestyle='solid',label='Nintendo')
act = sales_df[sales_df['Publisher'] == 'Electronic Arts'].groupby('Year')[['Global_Sales']].mean()
plt.plot(act['Global_Sales'], linestyle='dotted',label='EA')
act = sales_df[sales_df['Publisher'] == 'Activision'].groupby('Year')[['Global_Sales']].mean()
plt.plot(act['Global_Sales'], linestyle='dashed',label='Activision')
act = sales_df[sales_df['Publisher'] == 'Sony Computer Entertainment'].groupby('Year')[['Global_Sales']].mean()
plt.plot(act['Global_Sales'], linestyle='dashdot',label='Sony')
plt.legend()
```
